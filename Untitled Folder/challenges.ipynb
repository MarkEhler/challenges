### Fizzbuzz

```python
def fizzbuzz():
    # note that this function returns strings or integers
    for num in range(1,21):
        string = num
        if num % 3 == 0:
            string = 'fizz'
        if num % 5 == 0:
            try:
                string = string + 'buzz'
            except:
                string = 'buzz'
            else:
                string = 'fizzbuzz'
        print(string)
```

### Palendrome

```python
def palindrome(word):
    testing = list(map(list,word))
    reverse = testing[::-1]
    for idx, i in enumerate(testing):
        if testing[idx] != reverse[idx]:
            return "Not palindrome."
    return "Palindrome!"
```

```python
palindrome('test')
```

```python
palindrome('anna')
```

### Merging packages

if two pacakges in a list of packages meet the limit we merge them, if not we do nothing

```python
def get_indices_of_item_wights(arr, limit):
    m = list(map(lambda x: limit-x, arr))
    for idx, x in enumerate(arr):
        compliment = m[idx]
        if compliment in arr:
            comp_index = arr.index(compliment)
            return [idx,comp_index]
        
arr = [4, 6, 10, 15, 16]
lim = 21     
```

```python
get_indices_of_item_wights(arr, lim)
```

```python
def return_union(list_a, list_b):
    len1 = len(list_a)
    len2 = len(list_b)
    final_sorted_list = []
    j = 0
    k = 0

    for i in range(len1+len2):
        if k == len1:
            final_sorted_list.extend(list_b[j:])
            break
        elif j == len2:
            final_sorted_list.extend(list_a[k:])
            break
        elif list_a[k] < list_b[j]:
            final_sorted_list.append(list_a[k])
            k += 1
        else:
            final_sorted_list.append(list_b[j])
            j += 1
    return final_sorted_list
```
